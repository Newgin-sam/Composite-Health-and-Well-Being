```python
df = pd.read_csv('./OECD_social&civic_health_index_fil.csv')
```

```python
df.head(40)
```

```python
df = df.drop(columns=['STRUCTURE_ID','STRUCTURE_NAME','ACTION','Age', 'AGE','SEX','Sex','EDUCATION_LEV','Education level','DOMAIN','Time period','Observation value','OBS_STATUS','Observation status','UNIT_MULT','Unit multiplier','DECIMALS','Decimals','BASE_PER','Base period','UNIT_MEASURE',"Unit of measure", "Domain", "MEASURE"])
```

```python
df.head(40)
```

```python
pivot_df = df.pivot_table(index=["REF_AREA", "Reference area", "TIME_PERIOD"],
                            columns= "Measure",
                            values="OBS_VALUE")
pivot_df.reset_index(inplace=True)

pivot_df.head()
```

```python
pivot_df.shape
```

```python
pivot_df = pivot_df[pivot_df['TIME_PERIOD'] > 2009]
```
